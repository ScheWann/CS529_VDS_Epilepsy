```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# ep162 propagation data

# Load the combined JSON file
with open(
    "/Users/siyuanzhao/Documents/GitHub/CS529_VDS_Epilepsy/frontend/src/data/electrodes/ep162/ep162_all_network.json",
    "r",
) as file:
    combined_data = json.load(file)

# Iterate over each sample in the combined data
for sample_name, sample_data in combined_data.items():
    # Convert sample data to DataFrame
    propagation_df = pd.DataFrame(sample_data)

    electrode_counts = {}

    for network_list in propagation_df["network"]:
        if network_list:
            for item in network_list:
                source = item["source"]
                target = item["target"]

                # Increment source count
                if source in electrode_counts:
                    electrode_counts[source]["source_counts"] += 1
                else:
                    electrode_counts[source] = {
                        "electrode_id": source,
                        "source_counts": 1,
                        "target_counts": 0,
                    }

                # Increment target count
                if target in electrode_counts:
                    electrode_counts[target]["target_counts"] += 1
                else:
                    electrode_counts[target] = {
                        "electrode_id": target,
                        "source_counts": 0,
                        "target_counts": 1,
                    }

    # Convert to list
    result = list(electrode_counts.values())

    # Print result - optional, can remove if not needed
    print(f"{sample_name}: {result}")

    # Define output file name based on the sample name
    output_file = f"ep162_{sample_name}_propagation.json"
    # Save the result to a file
    with open(output_file, "w") as file:
        json.dump(result, file, indent=4)
```

```python
# New propagation data processing function

input_file_path = "/Users/siyuanzhao/Documents/GitHub/CS529_VDS_Epilepsy/frontend/src/data/electrodes/ep129/ep129_all_network.json"


def process_samples_from_file(input_file_path):
    with open(input_file_path, "r") as file:
        data = json.load(file)

    for sample_name, records in data.items():
        output_by_electrode = {}

        for record in records:
            electrodes = record.get("electrodes", [])
            network = record.get("network", [])
            network_exists = bool(network)

            for electrode in electrodes:
                if electrode not in output_by_electrode:
                    output_by_electrode[electrode] = {
                        "electrode_id": electrode,
                        "propagation": 0,
                        "nonPropagation": 0,
                        "sourceCount": 0,
                        "targetCount": 0,
                    }

                if network_exists:
                    output_by_electrode[electrode]["propagation"] += 1
                else:
                    output_by_electrode[electrode]["nonPropagation"] += 1

            if network_exists:
                for connection in network:
                    source = connection.get("source")
                    target = connection.get("target")

                    if source in output_by_electrode:
                        output_by_electrode[source]["sourceCount"] += 1
                    if target in output_by_electrode:
                        output_by_electrode[target]["targetCount"] += 1

        print(output_by_electrode)

        # Save
        result = list(output_by_electrode.values())
        output_file = f"ep129_{sample_name}_propagation.json"
        with open(output_file, "w") as file:
            json.dump(result, file, indent=4)


process_samples_from_file(input_file_path)
```

```python
electrode_df = pd.read_csv(
    "/Users/siyuanzhao/Documents/GitHub/CS529_VDS_Epilepsy/frontend/src/data/electrodes/ep164/ep164_rois_kmeans.csv"
)
electrode_df["E_Brain"] = electrode_df["E_Brain"].str.replace("\d+", "", regex=True)
electrode_df = electrode_df[electrode_df["E_Brain"] == "FP"]

electrode_numbers_set = set(electrode_df["electrode_number"].astype(int))


def initialize_output(electrode_numbers_set):
    output = {}
    for electrode in electrode_numbers_set:
        output[electrode] = {
            "electrode_id": electrode,
            "sourceCount": 0,
            "targetCount": 0,
            "nonPropagation": 0,
            "propagation": 0,
        }
    return output


def process_samples_from_file(input_file_path):
    with open(input_file_path, "r") as file:
        data = json.load(file)

    for sample_name, records in data.items():
        output_by_electrode = initialize_output(electrode_numbers_set)

        for record in records:
            network = record.get("network", [])
            for connection in network:
                source = connection.get("source")
                target = connection.get("target")
                if source in electrode_numbers_set and target in electrode_numbers_set:
                    output_by_electrode[source]["sourceCount"] += 1
                    output_by_electrode[target]["targetCount"] += 1

        result = list(output_by_electrode.values())

        output_file = f"ep164_{sample_name}_FP_electrode_counts.json"

        with open(output_file, "w") as file:
            json.dump(result, file, indent=4)


process_samples_from_file(input_file_path)
```
